 ### <span style='color:Red'> Unificar en un solo documento  </span>

```python
# outer join in python pandas

outer_join_df=pd.merge(taxonesaire, taxonessup, on=['taxon','taxlevel','dominio', 'phylo', 'clase', 'orden', 'familia',
       'genero'], how='outer')

outer_join_df.head(10)
```

```python
outer_join_df['total']= outer_join_df['total_x'] + outer_join_df['total_y']
```

```python
outer_join_df
```

Limpio el dataframe de columnas que se generan que no van a ser usadas en analisis posteriores

```python
del outer_join_df['rankID_x']
del outer_join_df['daughterlevels_x']
del outer_join_df['total_x']
del outer_join_df['rankID_y']
del outer_join_df['daughterlevels_y']
del outer_join_df['total_y']
```

```python
outer_join_df = outer_join_df[['taxlevel', 'taxon', 'EH-A', 'ICU-CO-A', 'ICU-TR-A', 'ICU-PR1-A',
       'OR-A', 'CO-A', 'TR-A', 'PR-A', 'ICU-PR2-A', 'ICU-PR2-S', 'ICU-CO-S',
       'ICU-TR-S', 'ICU-PR1-S', 'OR-S', 'EH-S', 'CO-S', 'TR-S', 'PR-S',
       'total', 'dominio', 'phylo',
       'clase', 'orden', 'familia', 'genero',]] 
```
