# KRONA

```python
tax6krona = df[df.taxlevel==6]
del tax6krona['taxlevel']
del tax6krona['taxon']
```

```python
tax6krona.columns
```

```python
levels = ['dominio', 'phylo', 'clase', 'orden', 'familia', 'genero']
```

```python
measures = [x for x in tax6krona if x not in levels]
measures
```

```python
for M in measures:
    columns = [M] + levels
    kromadf = tax6krona[columns]
    kromadf = kromadf[kromadf[M] > 0]
    destination = 'krona/%s.txt' % M
    kromadf.to_csv(destination, sep='\t', header=False, index=False)
```
