# COLUMSCREATOR
Este Notebook  genera columnas a partir de la información de las columnas rankID y taxon, para que sea posible unir varios documentos, sin perder la jerarquia del orden taxonomico con el que están generados, los documentos iniciales provienen de un análisis de secunecias con Mothur y la base de datos SILVA.


```python
import pandas as pd
import numpy as np
```

Cargamos el documento que sale del NOTEBOOK Uncultured_unclussified

```python
taxonesaire=pd.read_excel("Diversidad Hospital_UN_trated_Hpy.xlsx", sheet_name='AIRE')
```

```python
taxonesaire
```

```python
#taxonesaire['phylo'] = np.nan#creo unacolumna
```

```python
taxonesaire.head()
```

```python
fila = taxonesaire.iloc[3]
```

```python
fila
```

```python
lista_rankID= fila.rankID.split(sep='.')
```

```python

lista_rankID
```

```python
type(lista_rankID)
```

```python
len(lista_rankID)
```

```python
 level=2 # hago un contador, en este caso de esta fila quiero rellenar la primera columna con el filo bacteria que corresponde con 0.1 eso es nivel 2
```

```python
lista_selector=lista_rankID[0:level]
```

```python
 selector='.'.join(lista_selector)
```

```python
df = taxonesaire
```

```python
if len(lista_rankID[0:level]) < level:
     nombre_selector=np.nan
else:
    nombre_selector=df[df.rankID==selector].taxon 
    nombre_selector=nombre_selector.values[0]
```

```python
nombre_selector=df[df.rankID==selector].taxon #dataframe de una sola celda
nombre_selector=nombre_selector.values[0] #no es un dataframe es el valor de una celda
nombre_selector
```

```python
# Función para generar que van a contener las columnas quiero que cada fila contenga información 
```

```python
def column_creator(fila, level, df):
    """
    fila = 
    """
    try:
        lista_rankID= fila.rankID.split(sep='.')
        lista_selector=lista_rankID[0:level]
        selector='.'.join(lista_selector)
        nombre_selector=df[df.rankID==selector].taxon 
        if len(lista_rankID[0:level]) < level:
             nombre_selector=np.nan
        else:
            nombre_selector=nombre_selector.values[0]
    except Exception:
        nombre_selector=np.nan
    return nombre_selector

   
```

```python
column_creator(fila, 2, taxonesaire)
```

```python
taxonesaire.apply(column_creator,axis=1, args=(2,taxonesaire)) #aplicamos la funcion para rellenar la columna
```

```python
fila
```

```python
taxonesaire['dominio'] = taxonesaire.apply(column_creator,axis=1, args=(2,taxonesaire))
```

```python
taxonesaire['phylo'] = taxonesaire.apply(column_creator,axis=1, args=(3,taxonesaire))
```

```python
taxonesaire['clase'] = taxonesaire.apply(column_creator,axis=1, args=(4,taxonesaire))
```

```python
taxonesaire['orden'] = taxonesaire.apply(column_creator,axis=1, args=(5,taxonesaire))
```

```python
taxonesaire['familia'] = taxonesaire.apply(column_creator,axis=1, args=(6,taxonesaire))
```

```python
taxonesaire['genero'] = taxonesaire.apply(column_creator,axis=1, args=(7,taxonesaire))
```
