```python

from sklearn.cross_decomposition import CCA
```

```python
#from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import FunctionTransformer
```

```python
#metanormalized = StandardScaler().fit_transform(meta)
metanormalized = FunctionTransformer(np.log1p).fit_transform(meta)
```

```python
type(metanormalized)
```

```python
metanormalized.head(2)
```

```python
X_mc = tax6
X_mc.shape
```

```python
Y_mc = metanormalized
Y_mc.shape
```

```python
ca = CCA()
ca.fit(X_mc, Y_mc)
X_c, Y_c = ca.transform(X_mc, Y_mc)
```

```python
print(X_c.shape)
print(Y_c.shape)
```

```python
cc_res = pd.DataFrame({"CCX_1":X_c[:, 0],
                       "CCY_1":Y_c[:, 0],
                       "CCX_2":X_c[:, 1],
                       "CCY_2":Y_c[:, 1],
                       "Tª":metanormalized['Tª'],
                       "HR":metanormalized['H. relativa'],
                       "Partículas 0,3 micras":metanormalized['Particulas 0,3 m'],
                       "Chao-1":metanormalized['Chao-1'],
                       "Shannon H Index":metanormalized['Shannon H Index'],
                       "Simpson Index":metanormalized['Simpson Index']})
                       
                       
```

# KRONA

```python
tax6krona = df[df.taxlevel==6]
del tax6krona['taxlevel']
del tax6krona['taxon']
```

```python
tax6krona.columns
```

```python
levels = ['dominio', 'phylo', 'clase', 'orden', 'familia', 'genero']
```

```python
measures = [x for x in tax6krona if x not in levels]
measures
```

```python
for M in measures:
    columns = [M] + levels
    kromadf = tax6krona[columns]
    kromadf = kromadf[kromadf[M] > 0]
    destination = 'krona/%s.txt' % M
    kromadf.to_csv(destination, sep='\t', header=False, index=False)
```
